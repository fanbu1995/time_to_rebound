Read in the data worksheet

```python
dat = pd.read_excel("CA-DNA 022820_B2_CG.xlsx", sheet_name="B2 CA-DNA CG")
```

```python
dat.head()
```

```python
dat[dat.animal_ID=="RIm19"]
```

Go through each Animal ID, and extract all the measurements row by row

```python
All_IDs = dat.animal_ID.unique()
```

```python
data_dic = {'animal_id':All_IDs}
```

```python
for aid in All_IDs:
    dat_aid = dat[dat.animal_ID == aid]
    for i,row in dat_aid.iterrows():
        col_name = "DNA_copies_{}_{}".format(row['source'],row['weeks_infection'])
        if col_name not in data_dic:
            data_dic[col_name] = [row["SHIVgag_DNA_copies_per_million_CD4_tcells"]]
        else:
            data_dic[col_name].append(row["SHIVgag_DNA_copies_per_million_CD4_tcells"])
```

(check that for each column we have exactly 10 records)

```python
for col in data_dic:
    print(len(data_dic[col]))
```

Great! 10 entried per column.

Now make it into a dataframe and save it

```python
dat_DNA = pd.DataFrame(data_dic)
```

```python
dat_DNA.head()
```

```python
dat_DNA.to_csv("B2_CA_DNA.csv", index=False, index_label=False)
```
